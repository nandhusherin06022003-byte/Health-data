```python
df.drop_duplicates(inplace=True)
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

Column names were standardized to improve readability and avoid coding errors.

```python
df = df.convert_dtypes()
```

```python
df['record_id'] = range(1, len(df) + 1)
```

A derived column was created to uniquely identify each record.

```python
df_filtered = df.head(1000)
```

Used for faster analysis and visualization when dataset is large.

```python
df_summary = df.select_dtypes(include='number').mean()
df_summary
```

Aggregation was performed to understand average values of numerical health indicators.

```python
num_cols = df.select_dtypes(include='number').columns
cat_cols = df.select_dtypes(exclude='number').columns

num_cols, cat_cols
```

The dataset was divided into numerical and categorical columns for targeted analysis

```python
for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)

for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

Median and mode were used to preserve data distribution and avoid bias

```python
df.nunique()
```

Helps identify:

• Binary columns

• Low-variance columns

• Categorical features

```python
df = df.loc[:, df.nunique() > 1]
```

Columns with no variation were removed as they do not contribute to meaningful analysis

```python
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1

outliers = ((df[num_cols] < (Q1 - 1.5 * IQR)) | (df[num_cols] > (Q3 + 1.5 * IQR))).sum()
outliers
```

```python
df[num_cols] = (df[num_cols] - df[num_cols].min()) / (df[num_cols].max() - df[num_cols].min())
```

Numerical features were normalized to bring all values into a common range.

```python
df.to_csv("nfhs_final_preprocessed_data.csv", index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

These libraries are used to visualize patterns and trends in the data.

```python

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
exclude_keywords = ['id', 'code', 'no', 'index']

final_numeric_cols = [
    col for col in numeric_cols
    if not any(word in col.lower() for word in exclude_keywords)
]
final_numeric_cols
```

```python
df[final_numeric_cols[:3]].hist(figsize=(10,6), bins=10)
plt.suptitle("Distribution of Selected Numerical Indicators")
plt.tight_layout()
plt.show()
```

pop_f_6_sch → proportion of females with 6+ years of schooling

pop_below_15 → proportion of population below 15 years

sex_ratio_tot_pop → sex ratio of total population

This histogram shows the distribution of selected numerical health indicators in the dataset. It helps to understand how values are spread, identify skewness, and observe whether extreme values or outliers are present.

```python
plt.figure(figsize=(6,4))
plt.boxplot(df[final_numeric_cols[0]].dropna())
plt.title(f'Box Plot of {final_numeric_cols[0]}')
plt.ylabel(final_numeric_cols[0])
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(df[final_numeric_cols[1]].dropna())
plt.title(f'Box Plot of {final_numeric_cols[1]}')
plt.ylabel(final_numeric_cols[1])
plt.show()
```

This box plot shows the variation of another numerical indicator. Comparing multiple box plots helps understand differences in data spread and central tendency across variables

```python
plt.figure(figsize=(6,4))
plt.scatter(
    df[final_numeric_cols[0]],
    df[final_numeric_cols[1]],
    alpha=0.5
)
plt.xlabel(final_numeric_cols[0])
plt.ylabel(final_numeric_cols[1])
plt.title('Relationship Between Two Numerical Indicators')
plt.show()
```

This scatter plot shows the relationship between two numerical indicators. It helps identify patterns, trends, or correlations between variables

```python
plt.figure(figsize=(6,4))
plt.scatter(
    df[final_numeric_cols[1]],
    df[final_numeric_cols[2]],
    alpha=0.5,
)
plt.xlabel(final_numeric_cols[1])
plt.ylabel(final_numeric_cols[2])
plt.title('Relationship Between Numerical Indicators')
plt.show()
```

This scatter plot analyses another pair of numerical variables to understand whether a similar or different relationship exists. It helps strengthen exploratory analysis

```python
plt.figure(figsize=(8,4))
plt.plot(df[final_numeric_cols[0]].values)
plt.title(f'Trend of {final_numeric_cols[0]} Across Records')
plt.xlabel('Record Index')
plt.ylabel(final_numeric_cols[0])
plt.show()
```

This line plot represents how a numerical variable changes across dataset records. It helps observe overall trends, fluctuations, and stability of the variable.

```python
corr_data = df[final_numeric_cols[:5]].dropna()

plt.figure(figsize=(6,4))
sns.heatmap(corr_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

This heatmap displays the correlation between multiple numerical variables. It helps identify positive or negative relationships and the strength of association between variables.

```python
diabetes_prevalence = df[['wom_bld_sugar_vhigh_med', 'men_bld_sugar_vhigh_med']].mean().mean()
bp_prevalence = df[['wom_bp_ele_med', 'men_bp_ele_med']].mean().mean()

disease_prevalence = pd.Series({
    'Diabetes Prevalence': diabetes_prevalence,
    'BP Prevalence': bp_prevalence
})

disease_prevalence.plot(kind='bar', figsize=(6,4))
plt.title('Average Prevalence of Diseases')
plt.ylabel('Proportion')
plt.show()
```

This bar chart compares the average prevalence of major diseases across the population
